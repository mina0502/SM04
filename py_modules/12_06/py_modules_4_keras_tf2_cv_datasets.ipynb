### MNIST (Modified National Institute of Standards and Technology)
- (60000, 28, 28), train
- (10000, 28, 28), test
- grayscale images of handwritten single digits between 0 and 9

```
# use TensorFlow 2.x 
%tensorflow_version 2.x
# %tensorflow_version 1.x
```

```
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

print(tf.__version__)
```

```
from keras.datasets import mnist
```

```
# load dataset
(trainX, trainy), (testX, testy) = mnist.load_data()
# summarize loaded dataset
print('Train: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Test: X=%s, y=%s' % (testX.shape, testy.shape))
```

```
# plot first few images
for i in range(9):
	# define subplot
	plt.subplot(330 + 1 + i)
	# plot raw pixel data
	plt.imshow(trainX[i], cmap=plt.get_cmap('gray'))
# show the figure
# plt.show()
```

```
for i in range(9):
	# define subplot
	plt.subplot(250 + 1 + i)
	# plot raw pixel data
	plt.imshow(trainX[i], cmap=plt.get_cmap('gray'))
```

### CIFAR-10 Dataset (CIFAR: Canadian Institute For Advanced Research)
- (50000, 32, 32), train
- (10000, 32, 32) test
- color photographs of objects from 10 classes, such as frogs, birds, cats, ships, etc

```
from keras.datasets import cifar10
# load dataset
(trainX, trainy), (testX, testy) = cifar10.load_data()
# summarize loaded dataset
print('Train: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Test: X=%s, y=%s' % (testX.shape, testy.shape))
```

```
# plot first few images
for i in range(9):
	# define subplot
	plt.subplot(330 + 1 + i)
	# plot raw pixel data
	plt.imshow(trainX[i])
# show the figure
# plt.show()
```

### CIFAR-10 trainset

```
type(trainy)
unique, counts = np.unique(trainy, return_counts=True)
mnist_train_dic = dict(zip(unique,counts))
mnist_train_dic
```

```
plt.bar(list(mnist_train_dic.keys()),mnist_train_dic.values(),color='g')
```

```
bin = np.arange(-1, 11)
plt.hist(trainy, bin, alpha = 0.5, rwidth = 0.8)
plt.grid()
```

```
fig = plt.figure(figsize=(10,6))
for i in range(10):
  plt.subplot(2,5,i+1)
  num0_9 = trainX[trainy[:,0]==i]
  plt.imshow(num0_9[0])
  plt.title("Class %d" % (i))
  plt.xticks([])
  plt.yticks([])
```

### CIFAR-10 testset

```
type(testy)
unique, counts = np.unique(testy, return_counts=True)
mnist_test_dic = dict(zip(unique,counts))
mnist_test_dic
```

```
plt.bar(list(mnist_test_dic.keys()),mnist_test_dic.values(),color='g')
```
